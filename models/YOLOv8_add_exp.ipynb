```python
import os
import json
import pandas as pd
from ultralytics import YOLO

# 1. í´ë˜ìŠ¤ ID ë§¤í•‘ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°,
with open("/content/drive/MyDrive/codeit_level1_project/code/add_cls2cat_id_mapping.json", "r", encoding="utf-8") as f:
    classid_to_categoryid = json.load(f)

# 2. í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ íŒŒì¼ ì •ë ¬,
predict_dir = "/content/drive/MyDrive/codeit_level1_project/Data/test_images"
image_files = sorted(
    [f for f in os.listdir(predict_dir) if f.endswith((".png", ".jpg"))],
    key=lambda x: int(os.path.splitext(x)[0])
)

# 3. ê²°ê³¼ ì €ì¥ ë¦¬ìŠ¤íŠ¸,
csv_rows = []
annotation_id = 1

# 4. ì´ë¯¸ì§€ë³„ ì¶”ë¡ ,
for image_file in image_files:
    image_id = int(os.path.splitext(image_file)[0])  # ì˜ˆ: "3.png" â†’ 3
    img_path = os.path.join(predict_dir, image_file)
    results = model(img_path)

    for box in results[0].boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        bbox_w = x2 - x1
        bbox_h = y2 - y1

        cls_id = int(box.cls[0].item())
        conf = float(box.conf[0].item())
        category_id = int(classid_to_categoryid[str(cls_id)])  # str í‚¤ ì‚¬ìš© ì£¼ì˜

        csv_rows.append([
            annotation_id,
            image_id,
            category_id,
            round(x1),
            round(y1),
            round(bbox_w),
            round(bbox_h),
            round(conf, 2)
        ])

        annotation_id += 1

# 5. DataFrameìœ¼ë¡œ ë³€í™˜ ë° ì €ì¥ (ì»¬ëŸ¼ ìˆœì„œ ëª…ì‹œ),
columns = [
    "annotation_id", "image_id", "category_id",
    "bbox_x", "bbox_y", "bbox_w", "bbox_h", "score"
]
df = pd.DataFrame(csv_rows, columns=columns)
df.to_csv("submission_formatted.csv", index=False)
print("âœ… ì œì¶œìš© CSV ì €ì¥ ì™„ë£Œ: submission_formatted.csv")
```
